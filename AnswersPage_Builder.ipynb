Notebook works on the data taken from Quora. This is preparation notebook to build the QnA page using the answers.

1. Find the major topics among the questions
2. Group the questions and answer under the topics
3. Check for the answers that have considerable (above 100 views and 10 upvotes) and prioritize them on the top
4. Create multiple ways of looking at the questions and its answers and bring it on the first page of the site
5. Create a Repo to hold the pages and serve from there
6. Share the repo to everyone

HTML format file is created in the same folder.

```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

content = pd.read_csv('Concatenated.csv')
content.head(5)
```

```python
content.loc[content.Upvotes.isna(),'Upvotes'] = 0
content.loc[content.Upvotes == '–', 'Upvotes'] = 0
content.head()
```

```python
content.reset_index(inplace=True)
```

```python
content.drop(['Unnamed: 0','index'],inplace= True, axis= 1)
content.head()
```

```python
[float(vue.replace('K','')) for vue in content.Views if 'K' in vue]
```

```python
#function to convert and replace the K in the views
def rep_K(view: str):
    if "K" in view:
        view = view.replace('K','')
        view = float(view) * 1000

    else:
        view = float(view)

    return view
```

```python
content.Views = content.Views.apply(lambda x: rep_K(x))
```

```python
content.Upvotes = content.Upvotes.apply(lambda x: int(x))
```

```python
content.sort_values(by=['Upvotes','Views'],ascending = False, inplace= True,ignore_index=True)
```

```python
content.head()
```

```python
text_answer = content.Answers[2]
```

```python
#Creating the html file with the Questions and its answer in the html format. 

template = """
            </article>
        </section>   
    </body>
</html>
"""

with open('Answers_format.html','a',encoding="utf-8") as html:
    for ind, text in enumerate(text_answer.split('\n')):

        if ind %2 == 0:
            text = f"<p>{text}</p>\n"
            html.writelines(text)

        else:    
            text = f"<p>{text}</p>"
            html.writelines(text)

    html.write(template)
```
